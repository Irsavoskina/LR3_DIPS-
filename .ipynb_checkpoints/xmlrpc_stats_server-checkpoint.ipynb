#### Файл сервера xmlrpc_stats_server_p3.ipynb на Python 3 

```python

from xmlrpc.server import SimpleXMLRPCServer
from xmlrpc.server import SimpleXMLRPCRequestHandler

import datetime
import os
import glob

class RequestHandler(SimpleXMLRPCRequestHandler):
    rpc_paths = ('/RPC2',)

server = SimpleXMLRPCServer(("localhost", 8018), 
                            requestHandler=RequestHandler)
server.register_introspection_functions() 

max_count = 100
```

```python
#Сохранение старого файла с именем (дата последнего лога) и открытие нового
def renaming_old_log():  
    new_name = datetime.datetime.now().strftime("%Y%m%d_%H%M%S") + "_log.csv"
    os.rename('log.csv', new_name)
    f = open('log.csv','w')
    f.close()

#Добавление строки в лог
def add_log(sname):
    with open('log.csv','r') as f: 
        fn = f.readlines()
    if len(fn) >= max_count:
        renaming_old_log()
    with open('log.csv','a') as g:
        g.write(str(sname)+','+ datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S") +'\n')  
    return True

server.register_function(add_log, 'add_log')

    #Получение всех журналов из папки
    #def get_all_logs():
     #   log_files = glob.glob('*_log.csv')
     #   log_files.append('log.csv')
      #  return log_files   

   # path = "a/b/c"
   # filename = os.path.basename(path)
   # print(filename)  # Вывод: c

def get_log_files():
    log_files = glob.glob('*_log.csv')
    log_files.append('log.csv')
    return log_files 
       
    
def get_name_data_file():
    all_log_files = get_log_files()
    data_files = []
    for file in all_log_files:
        if len(file) > 8:
            data_time_file = (file[:15]).split("_")
            data_file = data_time_file[0]
            data_time = data_time_file[1]
            data = f"{data_file[:4]}-{data_file[4:6]}-{data_file[6:8]}"
            time = f"{data_time[:2]}:{data_time[2:4]}:{data_time[4:8]}"
            data_file = f"{data} {time}"
            data_files.append(data_file)

    return data_files    
#Получение журнала событий по времени/ событию и времени
def get_log(start_time, end_time, event = None):
    log = []
    files = []
    start_time = datetime.datetime.strptime(start_time, "%Y-%m-%d %H:%M:%S")
    end_time = datetime.datetime.strptime(end_time, "%Y-%m-%d %H:%M:%S")
    
    data_files = get_name_data_file()
    
    for data_file in data_files:
        data_file = datetime.datetime.strptime(data_file, "%Y-%m-%d %H:%M:%S")
        if data_file > start_time:
            names_file = (str(data_file)).replace("-","")
            name = names_file.replace(":","")
            name = name.replace(" ","_")
            name += "_log.csv"
            files.append(name)
        files.append('log.csv')
    
    #start_time = datetime.datetime.strptime(start_time, "%Y-%m-%d %H:%M:%S")
    #end_time = datetime.datetime.strptime(end_time, "%Y-%m-%d %H:%M:%S")
    
    for file in files:
        with open(file, 'r') as f:
            for line in f:
                log_line = line.strip().split(',')
                log_event, log_times = log_line
                log_time = datetime.datetime.strptime(log_times, "%Y-%m-%d %H:%M:%S")
                if (start_time <= log_time <= end_time ) and (event == log_event or event == None):
                    log.append(f"{log_time},{log_event}")
    return log
server.register_function(get_log, 'get_log')

# Получение журнала событий по событию 
def get_log_event(event):
    all_logs = get_log_files()
    log = []
    
    for file in all_logs:
        with open(file, 'r') as f:
            for line in f:
                log_line = line.strip().split(',')
                log_event, log_times = log_line
                if event == log_event:
                    log.append(f"{log_times},{log_event}")
    return log
                
server.register_function(get_log_event, 'get_log_event')


print ("Listening on port 8018...")
server.serve_forever()
```
